The purpose of this study was to compare the performance of Pymaceuticalsâ€™ drug of interest, Capomulin, against the other treatment regimens.Many statistical analysis methods and visualization tools has been used to understand the effect of  four drugs (Capomulin, Ramicane, Infubinol, and Ceftamin) and their effect on the tumor size.

.The regression analysis method helped to understand better that how much the average tumor volume (dependent variable) will change when weight of mice change(independent variables). The R-squared value is 0.70, that means 70% the model fit the data, that is a reliable value to predict the data from the model. Higher R-squared values displays smaller differences within the observed data, and the fitted value. Also 70% the model tells that all of the variation in the response variable around its mean.
.Hence from the selected treatments Capomulin and Ramicane reduces the volume of tumors better.


 

```python
# Dependencies and Setup
#%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Merge our two data frames together
combined_data = pd.merge(mouse_metadata,study_results , on="Mouse ID")

# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mouse_count = combined_data["Mouse ID"].nunique()
mouse_count
```

```python
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows = combined_data[combined_data.duplicated(subset=['Mouse ID', 'Timepoint'])]
#Get the unique list of Mouse IDs that have duplicate entries
duplicate_mice = duplicate_rows['Mouse ID'].unique()

# Print or work with the list of duplicate Mouse IDs
print(duplicate_mice)
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dublicate_data= combined_data[combined_data['Mouse ID'].isin(duplicate_mice)]
dublicate_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[combined_data['Mouse ID'].isin(duplicate_mice)==False]
clean_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_mouse = clean_data["Mouse ID"].nunique()
clean_mouse
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

mean_tumor_volume = clean_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
median_tumor_volume = clean_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
variance_tumor_volume =clean_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_dev_tumor_volume = clean_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
std_err_tumor_volume = clean_data.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary DataFrame.
summary_df = pd.DataFrame({
    "Mean Tumor Volume": mean_tumor_volume,
    "Median Tumor Volume": median_tumor_volume,
    "Variance Tumor Volume": variance_tumor_volume,
    "Tumor Volume Std.Dev.": std_dev_tumor_volume,
    "Tumor Volume Std.Err.": std_err_tumor_volume
})

# Print or work with the summary DataFrame
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
Capomulin_df = clean_data.loc[clean_data["Drug Regimen"] == "Capomulin",:]
Ramicane_df = clean_data.loc[clean_data["Drug Regimen"] == "Ramicane", :]
Infubinol_df = clean_data.loc[clean_data["Drug Regimen"] == "Infubinol", :]
Ceftamin_df = clean_data.loc[clean_data["Drug Regimen"] == "Ceftamin", :]

# Start by getting the last (greatest) timepoint for each mouse

Capomulin_final = Capomulin_df.groupby('Mouse ID').max()['Timepoint']
Capomulin_final_vol = pd.DataFrame(Capomulin_final)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
Capomulin_merge = pd.merge(Capomulin_final_vol, clean_data, on=("Mouse ID","Timepoint"),how="left")
Capomulin_merge.head()
```

```python
Capomulin_tumor = Capomulin_merge["Tumor Volume (mm3)"]

quartiles =Capomulin_tumor.quantile([.25,.5,.75])
lowerq = quartiles[0.25]
upperq = quartiles[0.75]
iqr = upperq-lowerq


print(f"The lower quartile of Capomulin tumors: {lowerq}")
print(f"The upper quartile of Capomulin tumors: {upperq}")
print(f"The interquartile range of Capomulin tumors: {iqr}")
print(f"The median of Capomulin tumors: {quartiles[0.5]} ")

lower_bound = lowerq - (1.5*iqr)
upper_bound = upperq + (1.5*iqr)

print(f"Values below {lower_bound} potential outliers.")
print(f"Values above {upper_bound} potential outliers.")
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list=["Capomulin", "Ramicane", "Infubinol","Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_volume_data=[]
treatment_df= combined_data[combined_data["Drug Regimen"].isin(treatment_list)]
treatment_df.head()

```

```python
# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = treatment_df.groupby(["Drug Regimen", "Mouse ID"]).agg(tumor_size=("Tumor Volume (mm3)", lambda x: x.iloc[-1]))
last_timepoint = last_timepoint.stack(level=0).unstack(level=0)
for treatments in treatment_list:
    print(treatments)
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    
drug = 0
for treatments in treatment_list:
    quartiles = last_timepoint[treatments].quantile([.25,.5,.75]).round(2)
    
    # Determine outliers using upper and lower bounds
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    
    lower_bound = round(lowerq - (1.5*iqr),2)
    upper_bound = round(upperq + (1.5*iqr),2)


    if drug == 0:
        print(f"------------------------------------------------------------")
    print(f"The lower quartile of {treatments} drugs is: {lowerq}")
    print(f"The upper quartile of {treatments} drugs is: {upperq}")
    print(f"The interquartile range of {treatments} drugs is: {iqr}")
    print(f"Values below {lower_bound} could be {treatments} outliers.")
    print(f"Values above {upper_bound} could be {treatments} outliers.")
    print(f"------------------------------------------------------------")
    drug+=1
   

    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

boxplot_list = []
for treatments in treatment_list:
    boxplot_list.append(list(last_timepoint[treatments].dropna()))
fig1, ax1 = plt.subplots()
ax1.set_title("Tumor Volume by Drug Regimen")
ax1.set_ylabel("Tumor Volume")
ax1.set_xlabel("Regimen")
ax1.boxplot(boxplot_list,labels=treatment_list)
plt.show()

    

    
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(capomulin_avg_vol['Weight (g)'],capomulin_avg_vol['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
#Regression
(slope, intercept,rvalue, pvalue, stderr)= linregress(capomulin_avg_vol["Weight (g)"],capomulin_avg_vol["Tumor Volume (mm3)"])
regress_values=capomulin_avg_vol["Weight (g)"]* slope + intercept
line_equation= f"y = {round(slope, 2)} x + {round(intercept, 2)}"

plt.scatter(capomulin_avg_vol["Weight (g)"],capomulin_avg_vol["Tumor Volume (mm3)"],color='b')
plt.plot(capomulin_avg_vol["Weight (g)"], regress_values, color='red')
plt.annotate(line_equation,(20,36), fontsize=14)
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Weight V Tumor Volume for Capomulin")
print(f"The r-squared is: {round(rvalue**2,3)}")
plt.show()
```
